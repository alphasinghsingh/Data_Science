```python

items = ("tv", "fridge", "wmachine", "shirts", "fans")
prices = (200, 300, 400, 200, 100)
discount = (10, 20, 30, 40, 20)

# -------------------------
# STEP 1 : Convert to lists
# -------------------------
items_list = list(items)
prices_list = list(prices)
discount_list = list(discount)

# -------------------------------------
# STEP 2 : All calculations on the lists
# -------------------------------------

total_bill = sum(prices_list)

discount_amounts_list = []
for i in range(len(prices_list)):
    dis_amt = prices_list[i] * discount_list[i] / 100
    discount_amounts_list.append(dis_amt)

total_discount = sum(discount_amounts_list)
payable_amount = total_bill - total_discount

# ------------------------------------------------
# STEP 3 : Convert lists → tuples for output display
# ------------------------------------------------
items_tuple = tuple(items_list)
prices_tuple = tuple(prices_list)
discount_tuple = tuple(discount_list)
discount_amounts_tuple = tuple(discount_amounts_list)

# -------------------------
# STEP 4 : Final output
# -------------------------
print("Items:", items_tuple)
print("Prices:", prices_tuple)
print("Discount %:", discount_tuple)
print("Discount Amounts:", discount_amounts_tuple)

print("\nTotal Bill =", total_bill)
print("Total Discount =", total_discount)
print("Final Payable Amount =", payable_amount)
```

```python
Menu = {
    "products": ["x", "y", "z", "p"],
    "price": [200, 300, 400, 500],
    "discount": [10, 20, 30, 40]
}

total_bill = 0
total_discount = 0

for i in range(len(Menu["products"])):
    price = Menu["price"][i]
    discount_percent = Menu["discount"][i]

    total_bill += price
    total_discount += price * (discount_percent / 100)
final_payable = total_bill - total_discount

print("Total Bill:", total_bill)
print("Total Discount:", total_discount)
print("Final Payable Amount:", final_payable)

```

```python
# Given dictionary
items = {
    "state": ["pb", "hr", "up", "hp", "pb", "guj", "hr"],
    "sale":  [2000, 3000, 4000, 5000, 1000, 3000, 2000]
}

total_sale = 0
state_wise_sale = {}


for i in range(len(items["state"])):
    state = items["state"][i]
    sale_amount = items["sale"][i]


    total_sale += sale_amount


    if state in state_wise_sale:
        state_wise_sale[state] += sale_amount
    else:
        state_wise_sale[state] = sale_amount


print("Total Sale:", total_sale)
print("State-wise Sale:", state_wise_sale)

```

```python
# Sum of squares of odd numbers from 1 to 9
sum_squares_odd = 0
for i in range(1, 10, 2):    # 1,3,5,7,9
    sum_squares_odd += i**2

# Sum of cubes of even numbers from 2 to 10
sum_cubes_even = 0
for i in range(2, 11, 2):    # 2,4,6,8,10
    sum_cubes_even += i**3

# Total sum of both series
total_sum = sum_squares_odd + sum_cubes_even

print("Sum of squares of odd numbers (1–9):", sum_squares_odd)
print("Sum of cubes of even numbers (2–10):", sum_cubes_even)
print("Total sum:", total_sum)
```

```python
import numpy as np

# Function to check if a number is prime
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Function to generate first 'count' prime numbers
def generate_primes(count):
    primes = []
    num = 2
    while len(primes) < count:
        if is_prime(num):
            primes.append(num)
        num += 1
    return primes

# -------------------------------
# 1D Array of Prime Numbers
# -------------------------------
primes_1d = np.array(generate_primes(10))
print("1D Array of Prime Numbers:")
print(primes_1d)

# -------------------------------
# 2D Array of Prime Numbers (4x5)
# -------------------------------
primes_2d = np.array(generate_primes(20)).reshape(4, 5)
print("\n2D Array of Prime Numbers:")
print(primes_2d)

# -------------------------------
# 3D Matrix of Prime Numbers (2x3x4)
# -------------------------------
primes_3d = np.array(generate_primes(24)).reshape(2, 3, 4)
print("\n3D Matrix of Prime Numbers:")
print(primes_3d)
```

```python
import csv

# File path (Colab)
file_path = "sales_data (1).csv"

# Dictionaries to store sales
product_sales = {}
state_sales = {}
gender_sales = {}

with open(file_path, 'r') as file:
    reader = csv.reader(file)

    # Skip header
    header = next(reader)

    # Read rows
    for row in reader:
        gender = row[2]
        state = row[3]
        product = row[4]
        amount = int(row[5])

        # Product-wise sales
        if product in product_sales:
            product_sales[product] += amount
        else:
            product_sales[product] = amount

        # State-wise sales
        if state in state_sales:
            state_sales[state] += amount
        else:
            state_sales[state] = amount

        # Gender-wise sales
        if gender in gender_sales:
            gender_sales[gender] += amount
        else:
            gender_sales[gender] = amount


# ================= OUTPUT =================

print("Unique Product vs Product-wise Sales")
for product, sale in product_sales.items():
    print(product, ":", sale)

print("\nUnique State vs State-wise Sales")
for state, sale in state_sales.items():
    print(state, ":", sale)

print("\nUnique Gender vs Gender-wise Sales")
for gender, sale in gender_sales.items():
    print(gender, ":", sale)
```

```python
import csv
import matplotlib.pyplot as plt

file_path = "sales_data (1).csv"

# Dictionaries
product_sales = {}
state_sales = {}
gender_sales = {}

with open(file_path, 'r') as file:
    reader = csv.reader(file)

    # Skip header
    header = next(reader)

    for row in reader:
        gender = row[2]
        state = row[3]
        product = row[4]
        amount = int(row[5])

        # Product-wise sales
        product_sales[product] = product_sales.get(product, 0) + amount

        # State-wise sales
        state_sales[state] = state_sales.get(state, 0) + amount

        # Gender-wise sales
        gender_sales[gender] = gender_sales.get(gender, 0) + amount


# ================== GRAPH 1 ==================
# Product vs Product-wise Sale

plt.figure(figsize=(10,5))
plt.bar(product_sales.keys(), product_sales.values(), color='skyblue')
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.title("Product vs Product-wise Sales")
plt.xticks(rotation=45)
plt.show()


# ================== GRAPH 2 ==================
# State vs State-wise Sale

plt.figure(figsize=(10,5))
plt.bar(state_sales.keys(), state_sales.values(), color='orange')
plt.xlabel("State")
plt.ylabel("Total Sales")
plt.title("State vs State-wise Sales")
plt.xticks(rotation=45)
plt.show()


# ================== GRAPH 3 ==================
# Gender vs Gender-wise Sale

plt.figure(figsize=(6,4))
plt.bar(gender_sales.keys(), gender_sales.values(), color='green')
plt.xlabel("Gender")
plt.ylabel("Total Sales")
plt.title("Gender vs Gender-wise Sales")
plt.show()
```

```python
import csv

# lists to store unique values
unique_products = []
unique_states = []
unique_genders = []

with open("sales.csv", "r") as file:
    reader = csv.reader(file)
    header = next(reader)   # skip header

    for row in reader:
        gender = row[2]
        state = row[3]
        product = row[4]

        # check uniqueness manually
        if gender not in unique_genders:
            unique_genders.append(gender)

        if state not in unique_states:
            unique_states.append(state)

        if product not in unique_products:
            unique_products.append(product)

print("Unique Genders:", unique_genders)
print("Count of Unique Genders:", len(unique_genders))

print("\nUnique States:", unique_states)
print("Count of Unique States:", len(unique_states))

print("\nUnique Products:", unique_products)
print("Count of Unique Products:", len(unique_products))
```
