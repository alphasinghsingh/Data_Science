```python

items = ("tv", "fridge", "wmachine", "shirts", "fans")
prices = (200, 300, 400, 200, 100)
discount = (10, 20, 30, 40, 20)

# -------------------------
# STEP 1 : Convert to lists
# -------------------------
items_list = list(items)
prices_list = list(prices)
discount_list = list(discount)

# -------------------------------------
# STEP 2 : All calculations on the lists
# -------------------------------------

total_bill = sum(prices_list)

discount_amounts_list = []
for i in range(len(prices_list)):
    dis_amt = prices_list[i] * discount_list[i] / 100
    discount_amounts_list.append(dis_amt)

total_discount = sum(discount_amounts_list)
payable_amount = total_bill - total_discount

# ------------------------------------------------
# STEP 3 : Convert lists → tuples for output display
# ------------------------------------------------
items_tuple = tuple(items_list)
prices_tuple = tuple(prices_list)
discount_tuple = tuple(discount_list)
discount_amounts_tuple = tuple(discount_amounts_list)

# -------------------------
# STEP 4 : Final output
# -------------------------
print("Items:", items_tuple)
print("Prices:", prices_tuple)
print("Discount %:", discount_tuple)
print("Discount Amounts:", discount_amounts_tuple)

print("\nTotal Bill =", total_bill)
print("Total Discount =", total_discount)
print("Final Payable Amount =", payable_amount)
```

```python
Menu = {
    "products": ["x", "y", "z", "p"],
    "price": [200, 300, 400, 500],
    "discount": [10, 20, 30, 40]
}

total_bill = 0
total_discount = 0

for i in range(len(Menu["products"])):
    price = Menu["price"][i]
    discount_percent = Menu["discount"][i]

    total_bill += price
    total_discount += price * (discount_percent / 100)
final_payable = total_bill - total_discount

print("Total Bill:", total_bill)
print("Total Discount:", total_discount)
print("Final Payable Amount:", final_payable)

```

```python
# Given dictionary
items = {
    "state": ["pb", "hr", "up", "hp", "pb", "guj", "hr"],
    "sale":  [2000, 3000, 4000, 5000, 1000, 3000, 2000]
}

total_sale = 0
state_wise_sale = {}


for i in range(len(items["state"])):
    state = items["state"][i]
    sale_amount = items["sale"][i]


    total_sale += sale_amount


    if state in state_wise_sale:
        state_wise_sale[state] += sale_amount
    else:
        state_wise_sale[state] = sale_amount


print("Total Sale:", total_sale)
print("State-wise Sale:", state_wise_sale)

```
