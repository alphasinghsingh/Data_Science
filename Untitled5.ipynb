```python
def add_num(*args):
  sum=0
  for i in args:

    sum=sum+ i
  return(sum)
print(add_num(20,30,49))
```

```python
def factorial(n):
  if n==0 :
    return 1
  else:
    p= n*factorial(n-1)
  return p
num=int(input("\n enter the number:"))
fact= factorial(num)
print("\n factorial of a number is:",fact)
```

```python
def fibbo(n):

  if n<=1:
       return n
  else:
    return fibbo(n-1)+fibbo(n-2)

n= int(input(" enter the terms upto which number is desired:"))
fibbo_ser=list(map(fibbo,range(n)))
print(fibbo_ser)
```
