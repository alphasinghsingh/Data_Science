```python
sum=0
fact=1
n=int(input("\n enter the number upto which series of factorial sum is desired=:"))
for i in range(1, n+1) :
  fact = fact * i
  sum = sum + fact
print("\n sum",sum)

```

```python
#prime number
n =int(input("\n enter the number whose prime number is to be find:"))

for i in range(2, n):
  if n%i== 0 :
    print("\n number is not a prime")
    break
else :
  print("\n number is prime")
```
