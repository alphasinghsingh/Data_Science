```python
import math

# Function to compute LCM of two numbers
def lcm(a, b):
    return abs(a * b) // math.gcd(a, b)

# Function to compute LCM of three numbers
def lcm_three(a, b, c):
    return lcm(lcm(a, b), c)

# Function to compute HCF (GCD) of three numbers
def hcf_three(a, b, c):
    return math.gcd(math.gcd(a, b), c)

# Input
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

# Calculation
hcf = hcf_three(a, b, c)
lcm_value = lcm_three(a, b, c)

# Output
print("HCF (GCD) of the three numbers:", hcf)
print("LCM of the three numbers:", lcm_value)
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

# Input from user
num = int(input("Enter the number of terms: "))

print("Fibonacci Series:")
for i in range(num):
    print(fibonacci(i), end=" ")
```

```python
# Input from user
text = input("Enter a word or number: ")

# Check palindrome using inbuilt function 'reversed'
if text == "".join(reversed(text)):
    print(text, "is a palindrome")
else:
    print(text, "is not a palindrome")
```

```python
# Input from user
n = int(input("Enter a number: "))

# Using inbuilt functions sum() and range()
total_sum = sum(range(1, n + 1))
average = total_sum / n

print("Sum of first", n, "natural numbers:", total_sum)
print("Average of first", n, "natural numbers:", average)
```

```python
def is_prime(num):
    if num < 2:
        return False
    return all(num % i != 0 for i in range(2, int(num**0.5) + 1))

n = int(input("Enter n: "))
primes = list(filter(is_prime, range(1, n + 1)))

print("Prime numbers:", primes)
```

```python
num = input("Enter a number: ")
power = len(num)
total = sum(int(d)**power for d in num)

if total == int(num):
    print("Armstrong Number")
else:
    print("Not Armstrong Number")
```
