```python
# sum of n natural numbers:
n=int(input("\n enter the n number whose sum is desired:"))

sum =0
for i in range(1,n+1) :
  sum= sum+i
print("\n sum= " , sum)
```

```python
#factorial of number
n=int(input("\n enter the number whose factorial is required:"))
fact=1
for i in range(1, n+1, 1) :
  fact= fact * i
print("\n factorial is=",fact)
```
