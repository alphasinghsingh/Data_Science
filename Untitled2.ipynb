```python
coupon_codes = ["qaz", "wsx", "wa"]
coupon_per = [10, 20, 30]
pay_bill = 1000

coupon_code = input("Enter the coupon code: ")

if coupon_code in coupon_codes:
    index = coupon_codes.index(coupon_code)
    discount = pay_bill * coupon_per[index] / 100
    final_bill = pay_bill - discount
    print("Discount Applied:", coupon_per[index], "%")
    print("Final Bill:", final_bill)
else:
    print("Invalid Coupon Code")
```

```python
items, prices, discounts = [], [], []

n = int(input("Enter the total number of elements you want in each list: "))

for i in range(n):
    item = input(f"Enter {i+1}th item name: ")
    price = float(input(f"Enter price of {item}: "))
    discount = float(input(f"Enter discount (%) on {item}: "))

    items.append(item)
    prices.append(price)
    discounts.append(discount)

total_bill = 0
total_discount = 0

# Calculate totals
for i in range(n):
    total_bill += prices[i]                       # Sum of prices
    discount_amount = prices[i] * discounts[i] / 100
    total_discount += discount_amount             # Sum of discount amounts

payable_bill = total_bill - total_discount

# Print results
print("\n---------- BILL SUMMARY ----------")
print("Total Bill (before discount):", total_bill)
print("Total Discount:", total_discount)
print("Payable Bill:", payable_bill)
print("----------------------------------")
```
