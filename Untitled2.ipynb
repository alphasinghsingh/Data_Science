```python
#total billing system
items =["piza", "burger", "fries","toast", "sandwitch"]
price =[200,500,300,200,350]
discount =[10,20,30,40,50]
total_bill=0
total_dis =0

for i in range(len(items)) :
  print(items[i])
  total_bill += price[i]
print(total_bill)
dis = price[i] * discount[i]/100
#print(items[i])
total_dis = total_dis + dis
print(total_dis)
final_dis = total_bill - total_dis
print(final_dis)
```
