```python
#if else
n = int(input("\n enter the number:"))
if n>0 :
  print("number is positive")
elif n<0 :
    print("\n number is negative")
else :
  print("\n number is equal to zero")
```

```python
#even or odd
n =int(input("\n enter the number"))
if n%2 ==0 :
  print("\n number is even")
else :
  print("\n number is odd")
```
